```python
# example
ex_pgn = '1. d4 d5 2. c4 c6 3. e3 a6 4. Nf3 e5 5. cxd5 e4 6. Ne5 cxd5 7. Qa4+ Bd7 8. Nxd7 Nxd7 9. Nc3 Nf6 10. Qb3 Be7 11. Nxd5 Qa5+ 12. Nc3 O-O 13. Be2 b5 14. O-O Rad8 15. Bd2 Qc7 16. Rac1 Qd6 17. Qc2 Qe6 18. Nb1 Bd6 19. a3 Nb6 20. Qc6 Nfd5 21. Ba5 Rc8 22. Qb7 Qh6 23. h3 Nc4 24. Bxc4 bxc4 25. Qxd5 Rfd8 26. Qxe4 Rd7 27. Bc3 Re7 28. Qf3 Re6 29. Nd2 Rf6 30. Qg4 Re8 31. Ne4 Rg6 32. Qd7 Rf8 33. Nxd6 Rxd6 34. Qc7 Rg6 35. Qh2 Re8 36. d5 f6 37. d6 Rd8 38. Rfd1 1-0'

# Split the input string into tokens, removing the move numbers
import re

def split_pgn(pgn):
    # Remove move numbers, results, and annotations (such as !, ?, !!, ??, !?, ?!)
    cleaned_pgn = re.sub(r'1-0|0-1|1/2-1/2|\*|[!?+#]+|\*|{[^}]*}|[%]\w+,\s?-?\d+', '', pgn)
    
    # Split the cleaned string into tokens
    tokens = cleaned_pgn.split()
    
    return tokens
print(split_pgn(ex_pgn))

ex_with_illegal = '1. e4 e5 {Good opening move by White} 2. Nf3 Nc6 {Standard development} 3. Bb5 a6 %eval, -0.3 {Opening theory, aiming for Ruy Lopez} 4. Ba4 Nf6 5. O-O Be7 6. Re1 b5 {[%eval, 0.2]} 7. Bb3 d6 {[%eval, 0.1]} 8. c3 O-O 9. h3 Nb8 10. d4 exd4 11. cxd4 Nb8 {[%eval, -0.1]} 12. Nc3 Nbd7 13. Bg5 h6 14. Bh4 Re8 15. a3 {[%eval, 0.0]} 16. Qd3 c5 {[%eval, -0.3]} 17. Rad1 cxd4 18. Nxd4 Nc5 19. Qf3 Qb6 {[%eval, -0.5]} 20. Nd5 Nxd5 21. Bxd5 Bxd5 22. exd5 a5 23. Rxe7 Rxe7 24. Nf5 Rd7 25. Qg3 g6 26. Nxh6+ Kg7 27. Nf5+ Kf8 28. Qg5 Re8 29. Qh6+ Kg8 30. Qg7# {[%eval, 1.5]} 1-0'

print(split_pgn(ex_with_illegal))
    
```

```python
import json

# Example: all_moves is the set of unique moves collected
# Adding <PAD> as the first token with index 0
vocab_dict = {"<PAD>": 0}
vocab_dict.update({move: idx for idx, move in enumerate(sorted(all_moves), start=1)})

# Save the dictionary to a JSON file
with open("vocab_dict.json", "w") as f:
    json.dump(vocab_dict, f)

# To load the dictionary later
with open("vocab_dict.json", "r") as f:
    loaded_vocab_dict = json.load(f)

# Print the loaded dictionary to verify
print(loaded_vocab_dict)
```

```python
# Function to convert a list of moves to their corresponding numeric tokens
def tokenize_moves(moves_list):
    return [vocab_dict.get(move, vocab_dict["<PAD>"]) for move in moves_list]

# Apply the function to the 'AN_tok' column to create a new column 'AN_tokenized'
df['AN_tok_int'] = df['AN_tok'].apply(tokenize_moves)
```

```python
# Create a reverse mapping from numbers to move strings
reverse_vocab_dict = {v: k for k, v in vocab_dict.items()}

def detokenize_moves(token_list):
    """
    Converts a list of numerical tokens back into chess move strings using the vocabulary dictionary.

    Args:
        token_list (list of int): List of numerical tokens.
        vocab_dict (dict): Dictionary mapping chess move strings to integers.

    Returns:
        list of str: List of chess moves corresponding to the tokens.
    """
   
    
    # Map each token back to its corresponding move string
    return [reverse_vocab_dict.get(token, "<UNK>") for token in token_list]

count = 0
for idx, moves in enumerate(df['AN_tok_int']):
    if 2 in moves:
        count+=1
        print(idx)
        print(detokenize_moves(moves))
        
        break

        
```
